```python
import json
from dataprofiler import Data, Profiler
```

```python
data = Data("/content/Titanic Dataset.csv")
data.data.head(5)
```

```python
data.shape
```

```python
host = data.data.sort_values(by='cabin')
host.head(5)
```

```python
print(data.data.shape[0])
len(host[host['cabin']!=""])
```

```python
profile = Profiler(data)
report  = profile.report(report_options={"output_format":"pretty"})
print(json.dumps(report, indent=4))
```

```python
new_data = Data("/content/Titanic Dataset.csv")
profile.update_profile(new_data)
report  = profile.report(report_options={"output_format":"pretty"})
print(json.dumps(report, indent=4))
```
