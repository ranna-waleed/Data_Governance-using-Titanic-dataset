```python
import pandera as pa
from pandera import Column
import pandas as pd

data_df = pd.read_csv('/content/tittanic.csv.csv', encoding='utf-8')

schema = pa.DataFrameSchema(
    {
        "pclass": Column(int),  # Passenger class
        "survived": Column(int),  # Survival status
        "name": Column(str),  # Passenger name
        "sex": Column(str),  # Gender
        "age": Column(float, nullable=True),  # Allow null values for Age
        "sibsp": Column(int),  # Number of siblings/spouses aboard
        "parch": Column(int),  # Number of parents/children aboard
        "ticket": Column(str),  # Ticket number
        "fare": Column(float, nullable=True),  # Allow null values for Fare
        "cabin": Column(str, nullable=True),  # Allow null values for Cabin
        "embarked":Column(str, nullable=True)  # Port of embarkation
    }
)

# # Validate the DataFrame against the schema
# validated_df = schema(data_df)
# print(validated_df)
```

```python
try:
    schema.validate(data_df, lazy=True)
except pa.errors.SchemaErrors as exc:
    failure_cases_df = exc.failure_cases
    display(exc.failure_cases)
```

```python
!pip install Cerberus
from cerberus import Validator
import pandas as pd
from cerberus import Validator
import pandas as pd

dfvalid = pd.DataFrame({"ID": [1, 2, 3],
                        "Age": [30, 20, 47],
                        "Name": ["Rana", "Mariam", "Renad"]})

v = Validator()
v.schema = {'Name': {'type': 'string', 'minlength': 5},
            'Age': {'type': 'integer', 'min': 18, 'max': 65},
            'ID': {'required': True, 'type': 'number'}}

for _, row in dfvalid.iterrows():
    if v.validate(row.to_dict()):
        print(f"Data for row {row.name} is valid")
    else:
        print(f"Data for row {row.name} is invalid")
        print(v.errors)
```

```python
import pandas as pd
import random
import numpy as np
data_df = pd.read_csv('/content/tittanic.csv.csv')
data_df
```

```python
import string
def randomize_values(df):
    for column in df.columns:
        if df[column].dtype == 'O':
            df[column] = [''.join(random.choices(string.ascii_letters + string.digits, k=10))
            for _ in range(len(df))]
    return df
```

```python
data_df['UniqueID'] = range(100, 100 + len(data_df))
data_df
```

```python
import numpy as np
df_perturbation = pd.read_csv('/content/tittanic.csv.csv')

def add_noise(df_perturbation, column, std = None):
    if std == None:
        std = df_perturbation[column].std()
    withNoise = df_perturbation[column].add(np.random.normal(0, std, df_perturbation.shape[0]))
    copy = df_perturbation.copy()
    copy[column] = withNoise
    return copy

perturbation = add_noise(df_perturbation, 'fare', std=100)
perturbation
```

```python
df = pd.read_csv("/content/tittanic.csv.csv")
def scramble_data(data):
    data_list = list(data)
    random.shuffle(data_list)
    return ''.join(data_list)

df['Scrambled_sex_condition'] = df['sex'].apply(scramble_data)
print(df[['sex', 'Scrambled_sex_condition']].head())
```
